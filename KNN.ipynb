## Definição do Problema

### Dados
Os dados abaixo são referentes ao cadastro de clientes de uma empresa de logística de entrega de alimentação. Nesse cadastro, há quanto foi gasto pelo cliente nas últimas 4 compras e uma classificação de NPS: **Promotor**, **Neutro**, **Detrator**. O nosso intuito é, a partir das últimas compras de alguns clientes que já deu uma nota NPS, estimar esse calssificação para aqueles que ainda não estão classificado, afim de melhorar o atendimento desses novos clientes.

Os dados abaixo seguem o seguinte padrão:

[**CPF**: INT, **Classificação NPS**: STRING, **Valor das últimas compras feitas pelo cliente**: TUPLA

### Funções

```python
# Função que calcula a distancia entre duas listas, no caso deste programa, seria entre as tuplas das listas data e no_class
def distancia(ponto1, ponto2):
    soma = 0
    
    # Esse FOR percorre as tuplas, elemento por elemento e faz os cálculos com os números de índices iguais 
    for i in range(len(ponto1)):
        soma += (ponto2[i] - ponto1[i])**2
    return soma ** (1/2)

# Função para ordenar os resultados e retorna K menores valores
# Os parâmetros são uma lista de valores e o valor de k
def buscar_k_menores_valores(lista_valores, k):
    lista_ordenada = sorted(lista_valores)
    return lista_ordenada[:k]

# Função para comparar quantas vezes cada string aparece na lista
def calculo_moda(lista_nps=[]):
    # Aqui o programa conta a incidência de cada palavra
    quantidade_promotor = lista_nps.count("Promotor")
    quantidade_neutro = lista_nps.count("Neutro")
    quantidade_detrator = lista_nps.count("Detrator")

    # O conjunto de if e elif's verifica os valores recebidos e retorna o que mais aparece
    if quantidade_promotor >= quantidade_neutro and quantidade_promotor >= quantidade_detrator:
        return [quantidade_promotor, "Promotor"]
    elif quantidade_neutro > quantidade_promotor and quantidade_neutro >= quantidade_detrator:
        return [quantidade_neutro, "Neutro"]
    elif quantidade_detrator > quantidade_neutro and quantidade_detrator > quantidade_promotor:
        return [quantidade_detrator, "Detrator"]

# Função para mostrar uma lista
def show_matriz(matriz):
    for item_lista in matriz:
        print(item_lista)
```

```python
lista_distancia = []
lista_distancia_indice = []
lista_k_menores_valores = []
lista_indice = []
lista_nps = []
k = 5
lista_final = []
```

```python
# Faz uma cópia da lista no_class
no_class_final = no_class.copy()

# Esse FOR percorre toda a lista no_class
for i in range(len(no_class)):
    # Aqui o programa precisa zerar as listas a cada iteracao de no_class
    lista_distancia=[]
    lista_nps=[]
    lista_indice=[]

    # Esse FOR interno percorre toda a lista data
    for j in range(len(data)):
        # chama função distância 
        # 1º parâmetro no_class[i][2] = 1° linha e o índice que tem a tupla
        # 2º parâmetro data[j][2] = 1° linha e o índice que tem a tupla
        distancia_total = distancia(no_class[i][2],data[j][2])
        # Adiciona os valores das distâncias em uma lista chamada lista_distancia
        lista_distancia.append(distancia_total)
    
    # Após ter tudo calculado e em uma lista, chamo a função lista_k_menores_valores
    # Os parâmetros dessa função é a minha lista_distancia e atribuímos o K = 5
    # Quando se tem um número par de classes, o ideal é usar o valor de k com um número ímpar para evitar empate
    lista_k_menores_valores = buscar_k_menores_valores(lista_distancia, k)
    
    # Pega os k menores valores das distancias calculadas e procura na lista_distancia os índices
    # Preciso dos índices pra saber qual valor de NPS está atrelado ao valor
    # Aqui pego todos os índices e salva em uma lista chamada lista_indice
    for numero in lista_k_menores_valores:
        indice = lista_distancia.index(numero)
        lista_indice.append(indice)

    #Esse FOR percorre a lista_indice
    for indice_string in lista_indice:
        # Pega o valor de NPS na lista data de acordo com os índices da lista_indice 
        # Salva os resultados em lista_nps
        lista_nps.append(data[indice_string][1])
    
    # A lista_moda recebe a função do cálculo da moda
    # O cálculo da moda usa como parâmetro a lista_nps
    # Aqui o programa cria uma lista com a quantidade de vezes que a string com maior incidência se repete
    # e também essa string repetida
    lista_moda = calculo_moda(lista_nps)
    
    # Insere o valor de maior incidência em uma cópia da lista no_class
    # A cópia foi criada para que a lista original não fosse alterada e consequentemente não interferisse no código
    # O índice que deve ser adicionado será o 1 de cada lista dentro da lista
    no_class_final[i][1] = lista_moda[1]
    
# Printa a quantidade de elementos na lista final apenas para verificação
print(len(no_class_final))
# Separador dos prints
print()
# Imprime a lista cópia de no_class com os valores alterados
# Os nomes de maior incidência já estão adicionados na posição [1] de cada sublista
show_matriz(no_class_final)
```
